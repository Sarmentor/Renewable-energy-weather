## 2. Weather data

### Read the data

Next, we read the weather data for Germany in 2016 by reading the full csv file.

The data in the file contains the following:

* wind
    - v1: velocity [m/s] @ height h1 (2 meters above displacement height)
    - v2: velocity [m/s] @ height h2 (10 meters above displacement height)
    - v_50m: velocity [m/s] @ 50 meters above ground
    - h1: height above ground [m] (h1 = displacement height +2m)
    - h2: height above ground [m] (h2 = displacement height +10m)
    - z0: roughness length [m]
* solar parameters:
    - SWTDN: total top-of-the-atmosphere horizontal radiation [W/mÂ²]
    - SWGDN: total ground horizontal radiation [W/mÂ²]
* temperature data
    - T: Temperature [K] @ 2 meters above displacement height (see h1)
* air data
    - Rho: air density [kg/mÂ³] @ surface
    - p: air pressure [Pa] @ surface

```python
weather = pd.read_csv("data/weather_data_GER_2016.csv",
                     parse_dates=[0], index_col=0)
```

```python
weather.head(3)
```

```python
weather.tail(3)
```

```python
weather.info()
```

Note that the data is given by chuncks of the German geographical territory, each determined by its latitute ('lat') and longitute ('lon'). Each chunck corresponds to the resolution of the MERRA-2 grid (which was used to obtain the data), which is 0.5Âº x 0.625Âº.

For example, the data for the time 2016-01-01 00:00:00 is the following:

```python
weather.loc[weather.index == '2016-01-01 00:00:00', :]
```

Therefore, for any given hour, we have data for 256 'chuncks' of Germany. Dividing the total number of entries by the number of 'chuncks' returns the expected number of hours of a leap year.

```python
2248704/256
```

Averaging over all the 'chuncks':

```python
weather_by_day = weather.groupby(weather.index).mean()
```

```python
weather_by_day.head(24)
```

## Algorithms

```python
combined.head(3)
```

```python
X_wind = combined[['v1', 'v2', 'v_50m', 'z0']]
y_wind = combined['DE_wind_generation_actual']

X_solar = combined[['SWTDN', 'SWGDN', 'T']]
y_solar = combined['DE_solar_generation_actual']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

### Linear regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
scores_wind = cross_val_score(lr, X_wind, y_wind, cv=5)
print(scores_wind, "\naverage =", np.mean(scores_wind))
```

```python
scores_solar = cross_val_score(lr, X_solar, y_solar, cv=5)
print(scores_solar, "\naverage =", np.mean(scores_solar))
```

### Ridge regression

```python
from sklearn.linear_model import Ridge
```

```python
for alpha in [0, 0.25, 0.5, 0.75, 1]:
    ridge = Ridge(alpha=alpha)
    scores_wind = cross_val_score(ridge, X_wind, y_wind, cv=5)
    print("alpha = ", alpha, "\n\tscores = ", scores_wind, "\n\taverage =", np.mean(scores_wind))
```

```python
for alpha in [0, 0.25, 0.5, 0.75, 1]:
    ridge = Ridge(alpha=alpha)
    scores_solar = cross_val_score(ridge, X_solar, y_solar, cv=5)
    print("alpha = ", alpha, "\n\tscores = ", scores_solar, "\n\taverage =", np.mean(scores_solar))
```

### Neural networks

```python
# Deep learning algorithms with keras
from keras.layers import Dense
from keras.models import Sequential
from keras import optimizers
```

```python
from keras import backend as K

def R2(y_true, y_pred):
    u = K.sum(K.square( y_true-y_pred ))
    v = K.sum(K.square( y_true - K.mean(y_true) ) )
    return ( 1 - u/(v + K.epsilon()) )
```

```python
X = combined.drop(['DE_wind_generation_actual', 'DE_solar_generation_actual', 'cumulated hours', 'lat', 'lon'], axis=1)
```

```python
X.head()
```

```python
X = X_wind
```

```python
from sklearn.preprocessing import scale
X_scaled = scale(X)
y_wind_scaled = scale(y_wind)
y_solar_scaled = scale(y_solar)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_scaled = sc.fit_transform(X)
y_wind_scaled = sc.fit_transform(y_wind.values.reshape(-1, 1))
y_solar_scaled = sc.fit_transform(y_solar.values.reshape(-1, 1))
```

```python
# Save predictors and its number to n_cols
n_cols = X_scaled.shape[1] 

# Set up the model
model = Sequential()

# Add the first layer
model.add(Dense(15, activation='relu', input_shape=(n_cols,), kernel_initializer='glorot_uniform'))

# Add the second layer
#model.add(Dense(20, activation='relu'))

# Add the output layer
model.add(Dense(1))

# Compile the model
adam = optimizers.Adam(lr=0.01)
model.compile(optimizer='adam', loss='mse', metrics=[R2])

# Fit the model
history = model.fit(X_scaled, y_wind_scaled, epochs=250, verbose=1, validation_split=0.3)

# Plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(history.history['loss'])
ax1.plot(history.history['val_loss'])
ax1.set_title('Mean squared error')
ax1.set_xlabel('Epoch')

ax2.plot(history.history['R2'])
ax2.plot(history.history['val_R2'])
ax2.set_title('$R^2$')
ax2.set_xlabel('Epoch')

plt.tight_layout()

# find predictions = {(prob of survival being False, prob of survival being True)}
#predictions = model.predict(y_wind)

# save predicted classes in y_pred
#y_pred = model.predict_classes(y_wind_test)

# score: returns the accuracy for the test data
#model.evaluate(X_wind_test, target_wind_test)
```
